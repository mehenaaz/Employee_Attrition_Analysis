```python
#import modules
import pandas as pd  # for dataframes
import matplotlib.pyplot as plt # for plotting graphs
import seaborn as sns # for plotting graphs
#% matplotlib inline
```

[https://www.datacamp.com/community/tutorials/predicting-employee-churn-python

```python
excelfile=pd.ExcelFile('Employee.xlsx')
```

## visualization of sheet 2(left employee)

```python
df1=excelfile.parse('Employees who have left')
print(df1)
```

```python
df1.info()
```

```python
df1.describe()
```

```python
# you can  plot a bar graph to count the number 
#of employees deployed on How many projects?
number_projects=df1.groupby('number_project').count()
plt.bar(number_projects.index.values, number_projects['satisfaction_level'])
plt.xlabel('Number of Projects')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#Similarly, you can also plot a bar graph to count the number 
#of employees have based on how much experience?
time_spent=df1.groupby('time_spend_company').count()
plt.bar(time_spent.index.values, time_spent['satisfaction_level'])
plt.xlabel('Number of Years Spend in Company')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#use Seaborn library and plot all the graphs in a single run using subplots#
features=['number_project','time_spend_company','Work_accident', 'promotion_last_5years','dept','salary']
fig=plt.subplots(figsize=(10,15))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace = 1.0)
    sns.countplot(x=j,data = df1)
    plt.xticks(rotation=90)
    plt.title("No. of employee")    
```

## visualization of sheet 1(current employee)

```python
df2=excelfile.parse('Existing employees')
print(df2)
```

```python
df2.info()
```

```python
df2.describe()
```

```python
features=['number_project','time_spend_company','Work_accident', 'promotion_last_5years','dept','salary']
fig=plt.subplots(figsize=(10,15))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace = 1.0)
    sns.countplot(x=j,data = df2)
    plt.xticks(rotation=90)
    plt.title("No. of employee") 
```

```python
#import module
from sklearn.cluster import KMeans
# Filter data
left_emp =  df1[['satisfaction_level', 'last_evaluation']]
# Create groups using K-means clustering.
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(left_emp)
```

```python
# Add new column "label" annd assign cluster labels.
left_emp['label'] = kmeans.labels_
# Draw scatter plot
plt.scatter(left_emp['satisfaction_level'], left_emp['last_evaluation'], c=left_emp['label'],cmap='Accent')
plt.xlabel('Satisfaction Level')
plt.ylabel('Last Evaluation')
plt.title('3 Clusters of employees who left')
plt.show()
```

```python
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
df2['salary']=le.fit_transform(df2['salary'])
df2['dept']=le.fit_transform(df2['dept'])
```
